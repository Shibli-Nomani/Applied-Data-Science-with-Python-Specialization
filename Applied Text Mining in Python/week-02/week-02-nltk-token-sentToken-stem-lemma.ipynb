<h2> 4. Normalization and Stemming </h2>

Normalization: Different forms of the same "word". Stremming is a process of choping the word and recover base word.

such as list, listing, listed, Listing, Listed [list]

<h3> step1: Normalize the Same Words</h3>

```python
input1 = "list List Listed listing Listings, lists"
```

<h5>Need to make it all lower case and splitting the words</h5>

```python
words1 = input1.lower().split(' ')
words1
```

<h3> step2: Stemming</h3>

```python
porter = nltk.PorterStemmer()
porter
```

```python
[porter.stem(t) for t in words1]
```

<h2>5. Lemmatization </h2>

It takes the words and break into a lemma.
example: walks, walking, wakled [lemmatization produces the lemma of those of group of word 'walk']



```python
udhr = nltk.corpus.udhr.words('English-Latin1') #universal declaration of Human Rights
```

```python
udhr
```

<h5>First 20 words</h5>

```python
udhr[:20]
```

<h5>stemming</h5>

```python
[porter.stem(t) for t in udhr[:20]]
```

**note:** some words are not valid after stemming

<h5>lemmatization works like stemming but preserving the core words and produces valid word.</h5>

**Meaningful Normalization**

```python
WNlemma = nltk.WordNetLemmatizer()
```

```python
[WNlemma.lemmatize(t) for t in udhr[:20]]
```

**note:** return valid words by investigaing some of the words randomly. here small 'rights' is lemmatized to 'right' since 'Rights' has different word with different meaning and it's not lemmatized
