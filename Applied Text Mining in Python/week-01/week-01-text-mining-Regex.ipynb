```python
words = [w for w in text2 if w.endswith('s')]
print('Words end with End s:', words)
```

# Regular Expression (regEx)

reference link of python for regular expression: https://docs.python.org/3/library/re.html

```python
import re
```

**`'[A-Za-z0-9_]'`:** Character from A-Z , a-z, 0-9

**`'@[A-Za-z0-9_]+'`:** we need to start with **@** and then follow the patterns **[A-Za-z0-9_]** with the **[]+** at least one or more time. **plus** matches one or more time with the pattern (alphabet, number, _ )

```python
import re
```

```python
text15 = 'Ethics are built right into the ideals and objectives of the United Nations #UNSG @ NY Society for Ethical Culture bit.ly/2guVelr @UN @UN_Women'
text15
```

# Callout Expression

```python
[w for w in text15.split() if re.search('@[A-Za-z0-9_]+', w)]
```

```python
[w for w in text15.split() if re.search('@\w+', w)]
```

<h3>here, \w is equivalent to [A-Za-z0-9_], `+` is repitition of same pattern in other words</h3>

```python
print('here, [\w] is equivalent to [A-Za-z0-9_], `+` is repitition of same pattern in other words')
```

<h2> Find Regular Expression of Data </h2>

```python
import re
```

```python
datestr = '23-10-2002\n23/10/2019\n15/07/21\n10/27/2020\n3/10/2019\n 17 Oct 1997\n24 October 2019\n 8 Oct 2020\nOctober 14, 2007\nOct 2 2022'
datestr
```

```python
re.findall(r'\d{2}[-/]\d{2}[-/]\d{4}', datestr)
```

```python
print('as we can see, above pattern will return some of the date not all')
```

```python
re.findall(r'\d{2}[-/]\d{2}[-/]\d{2,4}', datestr) #\d{2,4}: find the digit/number consists of 2 numeric values or 4
```

```python
re.findall(r'\d{2} (Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec) \d{4}', datestr) #(|) refers as OR
```

```python
re.findall(r'\d{2} (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec) \d{4}', datestr)
```

<h4> \d{1,2}: refers it as it 1 digit or 2 digit numeric </h4>
<h4> ?: refers it as it will include initial digits + Jan </h4>
<h4> [a-z] uses to match other words after Jan/Feb ('uary'/'ruary')</h4>
<h4> * refers as iteration</h4>

```python
re.findall(r'\d{1,2} (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{4}',datestr)
```

```python
re.findall(r'(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* (?:\d{1,2}, )?\d{4}',datestr)
```
