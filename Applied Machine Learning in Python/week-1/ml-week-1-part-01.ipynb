```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
%matplotlib inline

from sklearn.model_selection import train_test_split
```

```python
fruits= pd.read_table("../course3/assets/fruit_data_with_colors.txt")
fruits.head()
```

```python
fruits.info()
```

```python
fruits_label = pd.unique(fruits[['fruit_name', 'fruit_label']].values.ravel())
fruits_label
```

<h3>Alternative</h3>

```python
lookup_fruit_names = dict(zip(fruits.fruit_label.unique(), fruits.fruit_name.unique()))
lookup_fruit_names
```

```python
X= fruits[['mass', 'width', 'height', 'color_score']]
y = fruits['fruit_label']
```

<h3>train test splitting</h3>

```python
#by default 75/25
X_train, X_test, y_train, y_test =  train_test_split(X,y, random_state =0)
```

**Note:** It has been always recommanded to perform missing value treatement, data preprocessing and drop noise data before splitting

# Data Visualization

```python
from matplotlib import cm
from pandas.plotting._misc import scatter_matrix

cmap = cm.get_cmap('gnuplot')
scatter = scatter_matrix(X_train, c = y_train, marker = 'o', s = 40, hist_kwds = {'bins':15}, figsize = (12,12), cmap =cmap)
```

<h1>Three Dimensional Feature scatterplot</h1>

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X_train['width'], X_train['height'], X_train['color_score'],c= y_train, marker = 'o', s=100)
ax.set_xlabel('width')
ax.set_ylabel('height')
ax.set_zlabel('color_score')
plt.show()
```
