# Removing Text and Customization

```python
plt.figure(figsize = (10,4))
pos = nx.get_node_attributes(G, 'locations')
nx.draw_networkx(G, alpha = 0.4, with_labels = False, edge_color = '.4', node_color = 'Green')
plt.axis('off')
plt.tight_layout();
```

```python
plt.figure(figsize = (8,4))
pos = nx.get_node_attributes(G, 'location')
node_color = [G.degree(v) for v in G]
#node_size = [0.0005*nx.get_node_attributes(G, 'population')[v] for v in G]
#edge_width = [0.00015*G[u][v] ['weight'] for u, v in G.edges()]
```

```python
node_size = []
for v in G.nodes():
    population = G.nodes[v].get('population', 0)
    node_size.append(0.0005 * population)
```

```python
edge_width = []
for u, v, data in G.edges(data=True):
    weight = data.get('weight', 1)
    edge_width.append(0.00015 * weight)
```

```python
print("We can quickly which nodes has high population and high degrees.")
plt.figure(figsize = (10,8))
nx.draw_networkx(G, pos, node_size=node_size, 
                 node_color=node_color, alpha=1, with_labels=False, 
                 width=edge_width, edge_color='.4', cmap=plt.cm.Blues)

plt.axis('off')
plt.tight_layout()
```

# Draw Network where weight of edges is greater 770 using different color 

```python
plt.figure(figsize = (8,4))
pos = nx.get_node_attributes(G, 'location')
node_color = [G.degree(v) for v in G]
```

```python
node_size = []
for v in G.nodes():
    population = G.nodes[v].get('population', 0)
    node_size.append(0.0005 * population)
```

```python
edge_width = []
for u, v, data in G.edges(data=True):
    weight = data.get('weight', 1)
    edge_width.append(0.00015 * weight)
```

```python
print("We can quickly which nodes has high population and high degrees.")
plt.figure(figsize = (15,8))
nx.draw_networkx(G, pos, node_size=node_size, 
                 node_color=node_color, alpha=1, with_labels=False, 
                 width=edge_width, edge_color='.4', cmap=plt.cm.Blues)

greater_than_770 = [x for x in G.edges(data=True) if x[2]['weight'] > 770]
less_than_equal_770 = [x for x in G.edges(data=True) if x[2]['weight'] <= 770]
nx.draw_networkx_edges(G, pos, edgelist = greater_than_770 , alpha = 0.5, edge_color = 'red', width = 6) #alpha = opacity
nx.draw_networkx_edges(G, pos, edgelist = less_than_equal_770 , alpha = 0.5, edge_color = 'blue', width = 2)
nx.draw_networkx_labels(G, pos, labels = {"Los Angeles, CA": 'LA', 'New York, NY': 'NYC'}, font_size =18,
                       font_color = 'w')
plt.axis('off')
plt.tight_layout()
```
