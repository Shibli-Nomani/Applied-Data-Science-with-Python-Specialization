# NetworkX for social network analysis

<h2>Differnet Types of Network Graph</h2>




<h2>Undirected Network</h2>

Undirectional: A towards B and B towards A are same
nodes: A, B, C

edges: connection/ links/ tie between two nodes A and B

weight: A and B or B and A carrying same.

relation: between A and B

```python
import networkx as nx
```

```python
G = nx.Graph()
G.add_edge('A', 'B', weight = 6, relation = 'family')
G.add_edge('B', 'C', weight = 13, relation = 'friend')
G.edges()
```

```python
G.edges(data = True)
```

```python
G.edges(data = 'relation') #relation attributes
```

```python
G.edges(data = 'weight') #weight attributes
```

```python
G.edges['A', 'B']['weight'] #weight value of specific file
```

```python
G.edges['B', 'A']['weight']
```

<h2>Directed Network</h2>

directional: A towards B

nodes: A, B, C

edges: connection/ links/ tie between two nodes A and B

weight: number of times A sent email to B = 10

relation: between A and B


```python
G = nx.DiGraph()
G.add_edge('A', 'B', weight = 10, relation = 'friend')
G.add_edge('B', 'C', weight = 15, relation = 'family')
```

```python
G.edges()
```

```python
G.edges(data = 'relation')
```

```python
G.edges['A', 'B']
```

```python
print("Since it's an direted edge (A towards B). Therefore we can't retrieve (B towards A). Error generates as a result")
G.edges['B', 'A']
```

# Bipartite Graphs

all set of nodes connect with other set of nodes



```python
import networkx as nx
from networkx.algorithms import bipartite
```

```python
L = nx.Graph()

#set of nodes labelled as 0
L.add_nodes_from(['A', 'B', 'C', 'D'], bipartite = 0) #set 0 ==> L

#set of nodes labelled as 1
L.add_nodes_from([1,2,3], bipartite = 1) #set 1 ==> R
```

```python
L.add_edges_from([('A',1), ('A',2), ('B',2), ('C',1), ('C',3), ('D',3)])
```

```python
#to check if the B is Bipartite
bipartite.is_bipartite(B)
```

```python
L.add_edge('A', 'B')
```

```python
print("L is not bipartite any longer")
bipartite.is_bipartite(L)
```

```python
L.remove_edge('A', 'B')
```

```python
bipartite.is_bipartite(L)
```

<h3>separate the outcome from the dictionary value from outcomes col in pandas (using map and lambda)</h3>


```python
df['outcomes'] = df['outcomes'].map(lambda x: x['outcome'])
df.head()
```

```python
df['outcomes'].value_counts()
```

```python
won_as_white = df[df['outcomes']==1].groupby('white').sum()
won_as_white.head()
```

```python
won_as_black = df[df['outcomes']==-1].groupby('black').sum()
won_as_black.head()
```

```python
win_count = won_as_white.add(won_as_black, fill_value = 0)
win_count.head()
```

```python
win_count.nlargest(5, 'outcomes')
```
