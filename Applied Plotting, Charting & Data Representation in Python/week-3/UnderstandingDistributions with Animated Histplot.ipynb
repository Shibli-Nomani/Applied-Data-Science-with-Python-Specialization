```python
import matplotlib.pyplot as plt
import numpy as np
#for artist layer
%matplotlib notebook
import matplotlib.gridspec as gridspec

#for animation
import matplotlib.animation as animation

```

```python


#for plot figure
fig = plt.figure()



#x and y label positioning
fig.text(0.5, 0.04, 'Value-xlabel', ha='center', va='center')
fig.text(0.01, 0.5, 'Frequency-ylabel', ha='center', va='center', rotation='vertical')
def upadte_plots(curr):
    
    #no of iteration
    n= 1000
    if curr == n:
        a.event_source.stop() 
        
    #4 types of Ditstribution
    x1 = np.random.normal(-2.5, 1, 10000)
    x2 = np.random.gamma(2, 1.5, 10000)
    x3 = np.random.exponential(2, 10000)+7
    x4 = np.random.uniform(14,20, 10000)
    
    #2 by 2 sploting
        
    gspec = gridspec.GridSpec(2,2)
    gspec.update(wspace = 0.25, hspace = 0.5) #width and height space  between two subplot
    
    #subplotting placement

    top_left = plt.subplot(gspec[0,0])
    lower_left = plt.subplot(gspec[1:,0])
    top_right = plt.subplot(gspec[0,1:]) 
    lower_right = plt.subplot(gspec[1:, 1:])
    
   
    #histplot for 4 different distributions
    
    top_left.hist(x1[:curr], normed=True, bins=100, alpha=0.5,  fc="navy")
    top_left.title.set_text("Normal Distribution")
    lower_left.hist(x2[:curr], normed=True, bins=100, alpha=0.5,  fc="salmon" )
    lower_left.title.set_text("Gamma Distribution")
    top_right.hist(x3[:curr], normed=True, bins=20, alpha=0.5,  fc="olive" )
    top_right.title.set_text("Exponential Distribution")
    lower_right.hist(x4[:curr], normed=True, bins=100, alpha=0.5,  fc="forestgreen" )
    lower_right.title.set_text("Uniform Distribution")
    
    #return fig

#applied animation

a= animation.FuncAnimation(fig, upadte_plots, interval=2, blit=True, repeat=False) #frames=10
    
```
