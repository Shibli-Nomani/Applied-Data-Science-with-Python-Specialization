# Histogram

histogram is a bar chart to show the distribution of datapoints

1. Random 2. Uniform 3. Chi-Square Distribution

x-axis is giving distribution and y-axis is frequency of distribution

```python
fig, ((ax1,ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)

axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    
    sample = np.random.normal(loc =0.0, scale =1.0, size = sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

<h1>Clear Function</h1>

it will clear previous plotting and replace with new 

```python
import matplotlib.gridspec as gridspec
plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0,1:]) #all of the elemnts start with zero as we are using list
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])


#plotting data as per slicing

Y = np.random.normal(loc = 0.0, scale =1.0, size = 10000)
X = np.random.random(size=10000)

lower_right.scatter(X,Y)
top_histogram.hist(X,bins=100)
s = side_histogram.hist(Y, bins=100, orientation = 'horizontal')
```

```python
#clear the previous data and replace by new
top_histogram.clear()
top_histogram.hist(X, bins=100, normed = True)

side_histogram.clear()
side_histogram.hist(Y, bins =100, orientation = 'horizontal', normed = True)

side_histogram.invert_xaxis() #inversing the x-axis 
```

# To set the new range in axis and clear the axis

```python
# To clear it in more refine way

for ax in {top_histogram, lower_right}:
    ax.set_xlim(0,1)

for ax in [side_histogram, lower_right]:
    ax.set_ylim(-5,5)
```

# BOX PLot and Whisker Plot

a box plot / whisker plot is a mathod of showing min, max, median with Interquartile range as well as **outliers**

```python
import numpy as np
import pandas as pd

normal_sample = np.random.normal(loc=0.0, scale =1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size = 10000)

df = pd.DataFrame({'normal': normal_sample,
                  'random': random_sample,
                  'gamma': gamma_sample})

df.head()
```

```python
df.describe()
```

First Quantile = 25% of data

Second Quantile > 25% to = 75%

Third Quantile = 75%

Final = > 75% to = 100%

```python
plt.figure()

#underscore means unpacking value which we don't care about. We can avoid to show the value
_ = plt.boxplot(df['normal'], whis='range') 
```

# Multiple Box Plots

```python
plt.figure()
plt.clf()

_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis = 'range') #whis has been used to avoid outlier
```

# Animation

Matlab has animation and interactivity function. It happens using the backend

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
n = 100
X= np.random.randn(n)
```

Now we need create a function for animation which will call the frames within miliseconds start from frame 0

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla() #clear current object
    bins = np.arange(-4, 4, 0.5) #bound the bins -4 to 4with step 0.5
    plt.hist(X[:curr], bins = bins) #set axis value to stop auto scaling
    plt.axis([-4, 4, 0, 30]) #set axis value to stop auto scaling
    plt.gca().set_title('Sampling the normal distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    
    # place the text at certain position and show how many sample render to the screen
    plt.annotate('n = {}'.format(curr), [3,27]) 
    
```

```python
fig  = plt.figure()
#constructor = FuncAnimation
#1st Parameter = fig (figure that we want)
#2nd Parameter = update (function we define above)
#3rd Paramter = interval (amount of time we want betwwen update)



a = animation.FuncAnimation(fig, update, interval = 100)
```

# Interactivity

We need to inroduce canvas object which will handle all of the drawing events

types of hardware events: moving and clicking the pointer, pressing key of keyward

software events : java, html. piece of data is associated with function call. and it's an event

```python
plt.figure()
data = np.random.rand(10)

def oneclick(event):
    plt.cla() #clear current plot
    plt.plot(data)
    #create event
    plt.gca().set_title('Event at pixels {}, {} {}and data {},{}'.format(event.x,
                                                                        event.y,
                                                                        '\n',
                                                                        event.xdata,
                                                                        event.ydata))
#connect the event object with listener

plt.gcf().canvas.mpl_connect('button_press_event', oneclick) #event:button_press_event


    
    
```

```python
from random import shuffle
origins = ['China', 'Brazil', 'India', 'USA', 'Canada',
          'UK', 'Germany', 'Iraq', 'Chille', 'Maxico']

shuffle(origins)

df = pd.DataFrame({'Height': np.random.rand(10),
                  'Weight': np.random.rand(10),
                  'origin': origins})
df
```

```python
plt.figure()
# picker=5 means the mouse doesn't have to click directly on an event, but can be up to 5 pixels away
plt.scatter(df['Height'], df['Weight'], picker=5)
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))

# tell mpl_connect we want to pass a 'pick_event' into onpick when the event is detected
plt.gcf().canvas.mpl_connect('pick_event', onpick)
```
