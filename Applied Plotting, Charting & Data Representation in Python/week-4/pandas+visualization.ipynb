# Create DataFrame using numpy
Let's generate random number using random seeds and perform cumulative 

```python
np.random.seed(123)
df = pd.DataFrame({'A': np.random.randn(365).cumsum(0),
                  'B': np.random.randn(365).cumsum(0) + 20,
                  'C': np.random.randn(365).cumsum(0) - 10},
                  index = pd.date_range('1/1/2017', periods = 365))
                  
    
df.head()
```

# Plotting Data
Since we use plt style as a result visualization is quite different from normal condition
**plt.sytle.use('seaborn-whitegrid')**

```python
df.plot();
```

# Scatter Plot

```python
df.plot('A', 'B', kind = 'scatter')
```

```python
df.plot.scatter('A', 'B')
```

here we plot scatter for A against C with color and size change with B

```python
df.plot.scatter('A', 'C', c = 'B', s =df['B'], colormap = 'viridis')
```

# Setting Aspect Ratio

setting the aspect artio allows view to see that the range of 'A' is smaller than the range of 'C'

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap = 'viridis')
ax.set_aspect('equal')
```

# BOX Plot

```python
df.plot.box()
```

# Histogram

```python
df.plot.hist()
```

# Kernel Distribution Estimation(kde) Plot
It is used to visualize the probability density function of the continuous or non-parametric data variables

```python
df.plot.kde()
```

# pandas.tools.plotting

for high dimensional Data

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

# Scatter Matrix with Histogram for full dataset

It works parallelly between two columns in a dataframe

```python
# Scatter Matrix

pd.tools.plotting.scatter_matrix(iris);
```

# Parallel Dimension Plot
1. An easiest way to plot **high dimensional multi variant data** 
2. Each variable in a dataset has equally placed with parallel vertical line
3. Values of each variable have connected by line and different categories have been classified with different colors

```python
plt.figure()
plt.grid()
pd.tools.plotting.parallel_coordinates(iris, 'Name') #dataframe name and respective column name : 'Name'
```
