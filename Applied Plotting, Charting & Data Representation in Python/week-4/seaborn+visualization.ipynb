# Seaborn

1. Seaborn is a python visualiaztion libaries and works on top of matplotlib. It uses the matplotlib artist layer for visualization
2. Seaborn has enhanced functionality such as **Jointplot**, **Pairplot**, **Scatter Matrics Plot**, **Violin Plot**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
```

```python
np.random.seed(123)

v1 = pd.Series(np.random.normal(0,10,1000), name='v1') #start with 0, incremental 10 and sample 1000
v2 = pd.Series(np.random.normal(60,15,1000), name ='v2') #start with 60, incremental 15 and sample 1000
```

```python
bins=np.arange(-50,150,5)
bins
```

```python
plt.figure()
plt.hist(v1, alpha = 0.7, bins=np.arange(-50,150,5), label ='v1')
plt.hist(v2, alpha = 0.7, bins = np.arange(-50,150, 5), label ='v2')
plt.legend()
```

# Combination of Histogram and KDE Plot

Kernel Distribution Estimation works based on probabilty density of variable

```python
v3 = np.concatenate((v1,v2))
v3
```

```python
plt.figure()
plt.hist([v1,v2], histtype = 'barstacked', normed =True)
sns.kdeplot(v3)
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'navy'});
```

# Seaborn Joinplot
setting alpha = 0.4 helps us to see overlapping point

<h5> here we can see v1 and v2 are normally distributed variables. and positively correleted. In scatter plot, some points are overlapping each another.</h5>

```python
sns.jointplot(v1, v2, alpha =.4, color="g") 
```

# Seaborn plot tweaking using matplotlib

```python
grid = sns.jointplot(v1, v2, alpha =0.4, color = 'orange')
grid.ax_joint.set_aspect('equal')
```

# more way of jointplot visualization

```python
sns.jointplot(v1, v2, kind='hex')
```

```python
sns.set_style('white')
sns.jointplot(v1, v2, kind='kde', space=0, color = 'violet');
```

# Seaborn with categorical Data

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
g = sns.pairplot(iris, hue ='Name', diag_kind = 'kde')
g.fig.set_size_inches(9,7)
g._legend.set_bbox_to_anchor((.33, .94)); #default legend positioning

#here, we can see petal length and petal width are good options to separating the dataset
```

# Swarmplot(kind of scatter) and Violin plot (kind of box plot)

```python
plt.figure(figsize=(10,6))
plt.subplot(121) #row, col, 1st plot
sns.swarmplot('Name', 'PetalLength', data = iris)
plt.subplot(122)
sns.violinplot('Name', 'PetalLength', data = iris)
```
