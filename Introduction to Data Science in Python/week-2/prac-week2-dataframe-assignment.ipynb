# **`.extract()`**

there are 3 groups to retrive only two groups

`([\w]*)` : First Name

`(?:.*)` : **`?:`** is used not to retrun the value **`.*`** any type of whitespace

`([\w]*$)`: Last Name

```python
df.head(3)
```

```python
pattern = "(^[\w]*)(?:.* )([\w]*$)"
df['President'].str.extract(pattern).head()
```

# Identify pattern with column Name

**`"(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"`**

```python
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"
name = df['President'].str.extract(pattern).head()
name.head(3)
```

```python
df['First'] = name['First']
df['Last'] = name['Last']
```

```python
df.head(3)
```

**`""([\w]{3} [\w]{1,2}, [\w]{4})""`** 

[\w]{3} for Month **Feb** followed by space then [\w]{1,2} for date **22** use **`,`** then space followd by [\w]{4} for year **1732**

```python
df['Born'] = df['Born'].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
```

# D: 

**Correlation**

A correlation is a statistical relationship between two variables. If we wanted to know if vaccines work, we might look at the correlation between the use of the vaccine and whether it results in prevention of the infection or disease [1]. In this question, you are to see if there is a correlation between having had the chicken pox and the number of chickenpox vaccine doses given (varicella).

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
df = pd.read_csv("assignments/assignment2/assets/NISPUF17.csv")
```

```python
cpox_df = df[['HAD_CPOX', 'P_NUMVRC']]
cpox_df = cpox_df.dropna()
print(cpox_df['P_NUMVRC'].value_counts())
print(cpox_df['HAD_CPOX'].value_counts())

```

```python
#cpox_df = cpox_df[(cpox_df['P_NUMVRC']<3.0) & (cpox_df['P_NUMVRC']>0.0)]
cpox_df= cpox_df[(cpox_df.HAD_CPOX==1) | (cpox_df.HAD_CPOX==2)]
cpox_df['P_NUMVRC'].value_counts()
cpox_df['HAD_CPOX'].value_counts()
```

```python
cpox_df
```

```python
corr, pval=stats.pearsonr(cpox_df["HAD_CPOX"],cpox_df["P_NUMVRC"])
```

```python
corr
```
